## Series Title
- Manual check requires because anything can be title.
- By looking at each title, there was not an invalid or inappropriate entry.
- While checking, a duplicate was found with different released year.
    - The movie released later years have been renamed from 'Drishyam' to 'Drishyam 2' 

````python
# It's hard to judge and indicate if there are any title is inccorectly entered or not. 
# This requires a manual check.
newdf['Series_Title'].unique() 
newdf['Series_Title'].describe()

# Checking for duplicate entries in the Series_Title column.
duplicates = newdf['Series_Title'][newdf.duplicated('Series_Title', keep=False)]
# print(duplicates)
# Line 87 and 136 have the same Series_Title.

# Checking if they are completely the same entries or not.
newdf.loc[87]
newdf.loc[136]

# They are not the same entries. I will replace the Series_Title of line 136 with the correct title.
newdf.loc[136, 'Series_Title'] = 'Drishyam 2'

newdf['Series_Title'].describe()
# Now there are no duplicate entries in the Series_Title column.
````

## Runtime
- It is an object class with string and int values.
- Convert the type to string and remove string values.
- Change type to integer which is in minutes.
- Lastly, check to see if there is any suspicious entries, such as outliners.

````python
# Runtime is in object type and contains both int and string values.
newdf.sort_values(by='Runtime', ascending=True).head(5)

# - First I will convert the Runtime column to string.
newdf['Runtime'] = newdf['Runtime'].astype(str)

# - Then I will remove the 'min' from the string and convert it to integer.
newdf['Runtime'] = newdf['Runtime'].str.replace(' min', '', regex=True).astype(int)

# newdf.sort_values(by='Runtime', ascending=False).head(5)
newdf['Runtime'].describe() 

# Description of Runtime column looks fine. No outliners.
````

## Meta Score
- Missing 157 entries 
- Dropna may affect the entire analysis, hence replace missing values with average Meta Score

````python
# Checking the unique values in the Meta_score column.
newdf['Meta_score'].unique()

# See if incorrect entries are present in the missing Meta_score column.
newdf.loc[newdf['Meta_score'].isna()]

# Replace the missing values with the mean of the Meta_score column.
newdf['Meta_score'].fillna(newdf['Meta_score'].mean(), inplace=True)
newdf['Meta_score'].describe()
````
