```python
newdf.sort_values(by='Runtime', ascending=True).head(5)
# It seems that runtime is in minutes and data type is in object. 
# I will convert it to integer.

# - First I will convert the Runtime column to string.
newdf['Runtime'] = newdf['Runtime'].astype(str)

# - Then I will remove the 'min' from the string and convert it to integer.
newdf['Runtime'] = newdf['Runtime'].str.replace(' min', '', regex=True).astype(int)

# newdf.sort_values(by='Runtime', ascending=False).head(5)
newdf['Runtime'].describe() 
```

```python
# Totala missing Meta score is: 157
# The mean of the Meta score is: 77.0
# I will replace the missing values with the mean of the Meta score.
# Dropping the missing values is not a good idea because it will reduce the data size by 15%.
newdf['Meta_score'].unique()
newdf.loc[newdf['Meta_score'].isna()]
len(newdf.loc[newdf['Meta_score'].isna()])
newdf['Meta_score'].describe()
newdf['Meta_score'].fillna(newdf['Meta_score'].mean(), inplace=True)
newdf['Meta_score'].describe()
```
